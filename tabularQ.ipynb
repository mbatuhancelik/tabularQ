```python
import gym
import random
```

```python
#discrete values
dPosition = 3
dVelocity = 5
dAngle = 3
dAngularVelocity = 3
d = [dPosition, dVelocity, dAngle, dAngularVelocity]
```

```python
learningRate = 0.5
discountingFactor = 0.9
explorationFactor = 1
explorationDecay = 0.9995
```

```python
#populate states in the table
Q = {}
for i in range(dPosition+1):
    for k in range(dVelocity+1):
        for l in range(dAngle+1):
            for m in range(dAngularVelocity+1):
                for n in range(2):
                    Q[((i,k,l,m),n)] = 0
    
```

```python
numStates = len(Q.keys())
numStates
```

```python
def policy(state):
    global explorationFactor
    
    if random.random() < explorationFactor:
        return random.choice([0,1])
    elif Q[(state,0)] > Q[(state,1)]:
        return 0
    else:
        return 1
```

```python
env = gym.make("CartPole-v1")
env.reset()
```

```python
upperBound = env.observation_space.high
upperBound[1] = upperBound[0] * 2 
upperBound[3] = upperBound[2] * 2 
lowerBound =  env.observation_space.low
lowerBound[1] = lowerBound[0] * 2 
lowerBound[3] = lowerBound[2] * 2

valueRange = [0,0,0,0]
for i in range(len(upperBound)):
    valueRange[i] = upperBound[i] - lowerBound[i]

boundryPoints = [
    [lowerBound[k] + valueRange[k]/d[k] * i for i in range(1, d[k])] for k in range(4)
] 
for i in range(4):
    boundryPoints[i].insert(int((d[i]-1)/2), 0)
boundryPoints
```

```python
def quantisize(observation): 
    result = [dPosition  , dVelocity  , dAngle , dAngularVelocity ]
    for i in range(4):
        for k in range(len(boundryPoints[i])):
            if observation[i] < boundryPoints[i][k]:
                result[i] = k
                break
    return (result[0], result[1], result[2], result[3])
```

```python
def qmax(state):
    if Q[(state, 0)] > Q[(state, 1)]:
        return Q[(state, 0)]
    return Q[(state, 1)]
```

```python
 def singleRun():   
    state = quantisize(env.reset())
    global explorationFactor
#     print("initial condition:" , state)
    steps = 0
    for i in range(1000):
        action = policy(state)
        observation, reward, terminated, info = env.step(action)
        statePrime = quantisize(observation)
        target = 0
        
        if terminated:
            target = reward
            env.reset()
            # print(i, info)
            # break
        else:
            target = reward + discountingFactor * qmax(statePrime)
        Q[(state, action)] = (1-learningRate) *  Q[(state, action)] + learningRate * target
        state = statePrime
    
    env.close()
    return steps
```

```python
while explorationFactor > 10 ** -3:
    for i in range(10):
        singleRun()
    explorationFactor *= explorationDecay
    print(explorationFactor)
```
